```python
def region_specifics(big_list, list_type):
    southwest = 0
    southwest_count = 0
    northwest = 0
    northwest_count = 0
    southeast = 0
    southeast_count = 0
    northeast = 0
    northeast_count = 0
    for item in big_list:
        if item[0] == 'southwest':
            southwest += float(item[1])
            southwest_count += 1
        elif item[0] == 'northwest':
            northwest += float(item[1])
            northwest_count += 1
        elif item[0] == 'southeast':
            southeast += float(item[1])
            southeast_count += 1
        elif item[0] == 'northeast':
            northeast += float(item[1])
            northeast_count += 1
    return 'Southwest ' + list_type + ' average is: ' + str(southwest/southwest_count) + '. Northwest ' + list_type + ' average is: ' + str(northwest/northwest_count) + '. Southeast ' + list_type + ' average is: ' + str(southeast/southeast_count) + '. Northeast ' + list_type + ' average is: ' + str(northeast/northeast_count) + '.' 
```

```python
def region_specifics_word(big_list):
    southwest_male = 0
    southwest_female = 0
    southwest_count = 0
    northwest_male = 0
    northwest_female = 0
    northwest_count = 0
    southeast_male = 0
    southeast_female = 0
    southeast_count = 0
    northeast_male = 0
    northeast_female = 0
    northeast_count = 0
    for item in big_list:
        if item[0] == 'southwest':
            southwest_count += 1
            if item[1] == 'male':
                southwest_male += 1
            else:
                southwest_female += 1            
        elif item[0] == 'northwest':
            northwest_count += 1
            if item[1] == 'male':
                northwest_male += 1
            else:
                northwest_female += 1 
        elif item[0] == 'southeast':
            southeast_count += 1
            if item[1] == 'male':
                southeast_male += 1
            else:
                southeast_female += 1
        elif item[0] == 'northeast':
            northeast_count += 1
            if item[1] == 'male':
                northeast_male += 1
            else:
                northeast_female += 1
    return 'Southwest gender breakdown is ' + str((southwest_male/southwest_count)*100) + '% male and ' + str((southwest_female/southwest_count)*100) + '% female. Northwest gender breakdown is ' + str((northwest_male/northwest_count)*100) + '% male and ' + str((northwest_female/northwest_count)*100) + '% female. Southeast gender breakdown is ' + str((southeast_male/southeast_count)*100) + '% male and ' + str((southeast_female/southeast_count)*100) + '% female. Northeast gender breakdown is ' + str((northeast_male/northeast_count)*100) + '% male and ' + str((northeast_female/northeast_count)*100) + '% female.'
```

```python
def region_specifics_word(big_list):
    southwest_smoker = 0
    southwest_nonsmoker = 0
    southwest_count = 0
    northwest_smoker = 0
    northwest_nonsmoker = 0
    northwest_count = 0
    southeast_smoker = 0
    southeast_nonsmoker = 0
    southeast_count = 0
    northeast_smoker = 0
    northeast_nonsmoker = 0
    northeast_count = 0
    for item in big_list:
        if item[0] == 'southwest':
            southwest_count += 1
            if item[1] == 'yes':
                southwest_smoker += 1
            else:
                southwest_nonsmoker += 1            
        elif item[0] == 'northwest':
            northwest_count += 1
            if item[1] == 'yes':
                northwest_smoker += 1
            else:
                northwest_nonsmoker += 1 
        elif item[0] == 'southeast':
            southeast_count += 1
            if item[1] == 'yes':
                southeast_smoker += 1
            else:
                southeast_nonsmoker += 1
        elif item[0] == 'northeast':
            northeast_count += 1
            if item[1] == 'yes':
                northeast_smoker += 1
            else:
                northeast_nonsmoker += 1
    return 'Southwest smoker breakdown is ' + str((southwest_smoker/southwest_count)*100) + '% smoker and ' + str((southwest_nonsmoker/southwest_count)*100) + '% non-smoker. Northwest smoker breakdown is ' + str((northwest_smoker/northwest_count)*100) + '% smoker and ' + str((northwest_nonsmoker/northwest_count)*100) + '% non-smoker. Southeast smoker breakdown is ' + str((southeast_smoker/southeast_count)*100) + '% smoker and ' + str((southeast_nonsmoker/southeast_count)*100) + '% non-smoker. Northeast smoker breakdown is ' + str((northeast_smoker/northeast_count)*100) + '% smoker and ' + str((northeast_nonsmoker/northeast_count)*100) + '% non-smoker.'
```
